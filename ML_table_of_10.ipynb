# Problem Statement
Teach model to predict on the following equation


```
y = 10x
```

# Data Creation
Since the equation we want to teach our model is 

```
y = 10x
```
we need the following database


```
x = [0,10,20,30,40,50,...]
y = [0,10,20,30,40,50,...]
```






````python
x = [i for i in range(20+1)]
print(x)
````

````python
y= [i for i in range(10*20+1) if i%10==0] 
print(y)
````

# Approach 1
Non ML approach
Defining a function for the equation 
```
y = 10x
```



````python
def tableofTen(x):
  y = 10*x
  return y

for value in x:
    print(tableofTen(value))
````

# Approach 2
ML Approach

We got really excited to try ML out during the course so we ended up training a neuron on the table of 10, probably something that no one would ever use again,lol





````python
print(f'this is x:{x}')
print(f'this is y:{y}')
````

````python
xTrain = x[:-5] # Training data
yTrain = y[:-5] # Training data

xTest = x[-5:] # Testing data
yTest = y[-5:] # Testing labels

print(f'''
Training Data:
xTrain : {xTrain}
yTrain : {yTrain}

Testing Data:

xTest : {xTest}
yTest : {yTest}
''')
````

````python
#Importing the necessary libraries/frameworks

import tensorflow as tf
from tensorflow import keras
````

````python
#perceptron model
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1, input_shape=[1])
])
````

````python
model.compile(optimizer='adam', loss='mae')
````

````python
numEpoch = 5000 #Training it for 5000 times

model.fit(x=xTrain, y=yTrain, validation_data=(xTest, yTest), epochs=numEpoch)
````
